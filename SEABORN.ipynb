```python
#displot & histplot
#displot: represents the overall distribution of continuous data variables.
#hisplot:represents the distribution of one or more variables.
```

```python
df=sns.load_dataset('penguins')
df
```

```python
sns.displot(data=df, x='body_mass_g', kde=True, hue='sex', bins=10);
```

```python
plt.figure(figsize=(26,7))
sns.displot(data=df, x='body_mass_g', color='r', kde=True, hue='species', bins=10);
```

```python
sns.histplot(df, x='body_mass_g', color='r', lw=4, ls='-.', bins=20);#histplot bir veya daha fazla variable in dagilimini gosterir. renklendirme olabilir
```

```python
#kde:kernel density estimatee plot is a method for visualizing the distribution of observations in a dataset,
#represents data using a continues probability density curve
```

```python
sns.set(style='whitegrid')
sns.kdeplot(df.body_mass_g, clip=[2500,6000]);#2500 gramdan itibaren kesme islemi yapiyoruz. buna gore 2500 den 6000 gr a  kadar olan kisim alinmis oldu.
```

```python
sns.kdeplot(df.body_mass_g, clip=[2500,6000], bw_adjust=0.06, shade=True, color='y');#bw_adjust ck ayrintiya girmemizi sagladi ve cizginin sekli biraz bozuldu sanki mikroskopla bakmis gibi olduk.
```

```python
sns.kdeplot(df.body_mass_g, bw_adjust=3, shade=True, color='b');#simdi uzaydan bakiyoruz ayrintilar kayboldu.
```

```python
plt.figure(figsize=(10,6), dpi=500) #dpi kac tane nokta var demek
sns.kdeplot(df.body_mass_g, clip=[2000,6500], bw_adjust=2, shade=True, color='g');# uzaydan bakiyoruz ayrintilar kayip
```

```python
sns.boxplot(data=df, x='species', y='bill_depth_mm', hue='sex');
plt.legend(loc='best')
```

```python
plt.figure(figsize=(18,7), dpi=100)
sns.violinplot(data=df, x='species', y='bill_depth_mm', hue='sex',color='g', split=True);#split?
sns.boxplot(data=df,x='species', y='bill_depth_mm', hue='sex', color='r')
plt.legend(loc='best');
```

```python
plt.figure(figsize=(18,7), dpi=100);
sns.swarmplot(data=df, x='sex', y='bill_depth_mm', hue='species', split=True);
```

```python
plt.figure(figsize=(12,7));
sns.swarmplot(data=df, x='body_mass_g',y='sex',hue='species', size=7, dodge=True);
plt.legend(loc='best');
```

```python
plt.figure(figsize=(12,7));
sns.swarmplot(data=df, x='body_mass_g',y='sex',hue='species', size=7, dodge=False);# dodge =True denirse tur tur ayirir. yukaridaki grafige bak
plt.legend(loc='best');
```

```python
plt.figure(figsize=(18,7), dpi=100)
sns.boxenplot(data=df, y='body_mass_g',x='sex', hue='species', dodge=True);# daha ayrintili quartile leri gosterir.
plt.legend(loc='best')
```
